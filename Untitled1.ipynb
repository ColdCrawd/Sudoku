```python
import pandas as pd
import numpy as np
import random

#tablero_vacio = [[0 for _ in range(9)] for _ in range(9)]

sudoku_1f = [
   #este es toda la primera fila
        [1,2,3,4,5,6,7,8,9],
        [1,2,3,4,5,6,7,8,9],
        [1,2,3,4,5,6,7,8,9],
        [1,2,3,4,5,6,7,8,9],
        [1,2,3,4,5,6,7,8,9],
        [1,2,3,4,5,6,7,8,9],
        [1,2,3,4,5,6,7,8,9],
        [1,2,3,4,5,6,7,8,9],
        [1,2,3,4,5,6,7,8,9]
]

def crear_tablero():
    tablero_vacio = [[0 for _ in range(9)] for _ in range(9)]
    if generar_sudoku(tablero_vacio):
        return tablero_vacio
    return None

def tablero_medio(tablero):
    tablero_nuevo = [[0 for _ in range(9)] for _ in range(9)]

    posiciones = [(i, j) for i in range(9) for j in range(9)]
    random.shuffle(posiciones)

    cantidad = 40
    for fila, columna in posiciones[:cantidad]:
        tablero_nuevo[fila][columna] = tablero[fila][columna]

    return tablero_nuevo


def generar_sudoku(tablero):
    for i in range(9):
        for j in range(9):
          if(tablero[i][j] == 0):
            numero = random.sample(range(1, 10), 9)
            for n in numero:
                    if condicion_sudoku(tablero, i, j, n):
                        tablero[i][j] = n
                        if generar_sudoku(tablero):
                            return True
                        tablero[i][j] = 0
            return False
    return True


def condicion_sudoku(tablero, fila, columna, numero):
    for i in range(9):
        if tablero[fila][i] == numero or tablero[i][columna] == numero:
            return False

    inicio_fila = (fila // 3) * 3
    inicio_columna = (columna // 3) * 3
    for i in range(3):
        for j in range(3):
            if tablero[inicio_fila + i][inicio_columna + j] == numero:
                return False
    return True

def imprmir_tablero(sudoku):

    df = pd.DataFrame(sudoku, index=range(1, 10), columns=range(1, 10))

    tablero = ""
    for i in range(9):
        row = ""
        for j in range(9):
            row += f" {df.iloc[i, j] if df.iloc[i, j] != 0 else ' '} "
            if (j + 1) % 3 == 0 and j < 8:
                row += "|"
        tablero += row + "\n"
        if (i + 1) % 3 == 0 and i < 8:
            tablero += "-" * 30 + "\n"
    print(tablero)

def es_valido(tablero_completo, fila, columna, valor):
    if tablero_completo[fila, columna] == valor
        return True
    return False

def jugar(df):
    while True:
        imprmir_tablero(df)
        try:
            fila = int(input("Ingrese la fila (1-9): "))
            columna = int(input("Ingrese la columna (1-9): "))
            valor = int(input("Ingrese el valor (1-9): "))

            if fila < 0 or fila > 8 or columna < 0 or columna > 8 or valor < 1 or valor > 9:
                print("Entrada no valida. Asegurese de que la fila y columna esten entre 0 y 8, y el valor entre 1 y 9.")
                continue

            if es_valido(df, fila, columna, valor):
                df.iloc[fila, columna] = valor
                print(f"Se ha colocado el {valor} en la posicion ({fila}, {columna}).")
            else:
                print("Movimiento invalido. El valor no cumple con las reglas del Sudoku.")
        except ValueError:
            print("Entrada invalida. Intente de nuevo.")

        cont = input("¿Desea continuar jugando? (s/n): ").strip().lower()
        if cont != 's':
            break

def prueba():

  tablero_lleno = crear_tablero()

  if(tablero_lleno):
    imprmir_tablero(tablero_lleno)

    tablero_medio = tablero_medio(tablero_lleno)

    jugar(tablero_medio)

  else:
    print("No se pudo generar el sudoku")

def main():
  prueba()

if __name__ == "__main__":
    main()

```
